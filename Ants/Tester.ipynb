```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the pubs.csv file
df = pd.read_csv('pubs.csv')

# Extract the values of the OpeningTime column and add them to a list
ListOpeningTime = df['OpeningTime'].tolist()
ListClosingTime = df['ClosingTime'].tolist()
ListPopularity = df['Popularity'].tolist()

print(ListOpeningTime)
print(ListClosingTime)
print(ListPopularity)
```

```python
import numpy as np
import matplotlib.pyplot as plt

#create subplots (19 rows 1 column)
fig, axs = plt.subplots(19, 1, figsize=(10, 80))

for i in range(len(ListOpeningTime)):   
    currentTime = np.linspace(ListOpeningTime[i], ListClosingTime[i], ListClosingTime[i] - ListOpeningTime[i])
    queueMax = ListPopularity[i]
    queueLength = queueMax * np.exp(-0.1 * (currentTime - ListOpeningTime[i]))
    axs[i].plot(currentTime, queueLength)


plt.suptitle('Queuq length dependent on popularity and time',y=1.01)
plt.tight_layout()
plt.show()
```
